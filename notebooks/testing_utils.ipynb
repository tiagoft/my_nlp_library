```python
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import matplotlib.pyplot as plt
import my_nlp_library as nlp
```

```python
sample_size = 10_000
embedding_dim = 32

dataset_train, dataset_test, tokenizer, _ = nlp.get_imdb_dataset(sample_size=sample_size)
vocab_size = tokenizer.vocab_size

model = nlp.ExperimentModel(vocab_size=vocab_size, 
                            embedding_dim=embedding_dim, 
                            output_dim=1)

model.compile(optimizer=torch.optim.Adam(model.parameters(), lr=1e-3),
              device="cuda",
              metrics=["accuracy", 
                       "precision", 
                       "recall", 
                       "f1"])

history = model.fit(dataset_train, 
                   n_epochs=25,
                   validation_split=0.2)

print("Training finished")
```
